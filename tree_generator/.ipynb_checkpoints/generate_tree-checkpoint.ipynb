```python
import os
import sys
import glob
import h5py
import numpy as np
import matplotlib.pyplot as plt
sys.path.append(larpix_geometry_libpath)
sys.path.append(evd_libpath)
```

```python
raw_datapath = '/home/rberner/PACMAN/analysis/trees_livio/data/convertedData'
raw_filestring = 'datalog_2020_11_01_08_36_35_CET_evd.h5'
```

```python
plt.ion()
f = h5py.File(raw_datapath + str('/') + raw_filestring,'r')
```

```python
print(f.keys())
```

```python
trigger_types = []

counter_self_trig = 0
counter_external = 0
counter_cross = 0
counter_periodic = 0

for tr in f['packets'][:]['trigger_type']:
    if tr == 0:
        counter_self_trig += 1
    if tr == 1:
        counter_external += 1
    if tr == 2:
        counter_cross += 1
    if tr == 3:
        counter_periodic += 1

# print(f['packets'[f['packets']['packet_type']==7])

timestamps = []
timediff = []

for packet in f['packets'][:]:
    if packet['trigger_type']==2 and packet['packet_type'] == 7:
        timestamps.append(packet['timestamp'])
print(' first 10 timestamps: ', timestamps[0:10])

for x in range(len(timestamps)-1):
    thisdiff = timestamps[x+1] - timestamps[x]
    timediff.append(thisdiff)
    #if abs(thisdiff-1000000) > 1:
    #    print(thisdiff, type(thisdiff))
print(' first 10 timediffs:  ', timediff[0:10])

print(' n real triggers:   ', len(timestamps))
print(' counter_self_trig: ', counter_self_trig)
print(' counter_external:  ', counter_external)
print(' fraction:          ', counter_external/counter_self_trig)
print(' counter_cross:     ', counter_cross)
print(' counter_periodic:  ', counter_periodic)
print(' --------------------------------- ')
```
